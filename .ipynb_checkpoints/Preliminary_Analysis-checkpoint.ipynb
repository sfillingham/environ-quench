```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
centrals = pd.read_csv("data/hlist_1.00035_centralhalos.csv")
sats = pd.read_csv("data/hlist_1.00035_satellitehalos_1.0Rvir.csv")
```

```python
plt.figure(figsize=(8, 8))
plt.xlabel(r'$\rm Halo\ Mass\ (M_{vir})$')
sns.distplot(np.log10(centrals['mvir(10)']))
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.xlabel(r'$\rm Halo\ Mass\ (M_{vir})$')
sns.distplot(np.log10(sats['mvir(10)']))
plt.show()
```
